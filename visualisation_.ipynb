```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/content/creditScoreClassification.csv',low_memory=False)
```

#**OCCUPATIONAL COUNTS**

```python

occupation_counts = df['Occupation'].value_counts()
plt.bar(occupation_counts.index, occupation_counts.values,color = ['skyblue', 'lightgreen', 'lightcoral', 'lightpink', 'lightsalmon']
)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation Counts')
plt.xticks(rotation=45)


plt.show()
```

#**Counts of  the occurrences of each month**

```python
month_counts = df['Month'].value_counts()
colors= ['lightblue', 'blue', 'lightgreen', 'green', 'lightcoral',
          'red', 'lightyellow', 'orange', 'violet', 'purple']
plt.bar(month_counts.index, month_counts.values, color=colors[:len(month_counts)])
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Month Counts')

plt.show()
```

```python

credit_score_counts = df['Credit_Score'].value_counts()

colors = ['lightcoral', 'skyblue', 'lightgreen', 'salmon', 'plum']

plt.bar(credit_score_counts.index, credit_score_counts.values, color=colors[:len(credit_score_counts)])

plt.xticks(rotation=45)

plt.xlabel('Credit Score')
plt.ylabel('Count')
plt.title('Credit Score Counts')

plt.show()
```

```python

credit_mix_counts = df['Credit_Mix'].value_counts()

colors = ['lightblue', 'teal', 'darkcyan', 'steelblue', 'cadetblue']

plt.bar(credit_mix_counts.index, credit_mix_counts.values, color=colors[:len(credit_mix_counts)])

plt.xticks(rotation=45)

plt.xlabel('Credit Mix')
plt.ylabel('Count')
plt.title('Credit Mix Counts')

plt.show()
```

```python


data = pd.read_csv('/content/creditScoreClassification.csv')

# Convert columns with numerical data stored as objects to numeric (forcing errors to NaN where conversion fails)
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['Annual_Income'] = pd.to_numeric(data['Annual_Income'], errors='coerce')
data['Outstanding_Debt'] = pd.to_numeric(data['Outstanding_Debt'], errors='coerce')
data['Monthly_Balance'] = pd.to_numeric(data['Monthly_Balance'], errors='coerce')
data['Amount_invested_monthly'] = pd.to_numeric(data['Amount_invested_monthly'], errors='coerce')
data['Num_of_Delayed_Payment'] = pd.to_numeric(data['Num_of_Delayed_Payment'], errors='coerce')


numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numerical_data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
